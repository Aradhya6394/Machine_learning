# 3. Opening a csv file from an url

```python
import requests
import pandas as pd
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"

headers = {
    "User-Agent": "Mozilla/5.0"
}

req = requests.get(url, headers=headers)

data = StringIO(req.text)

df = pd.read_csv(data)
df
```

# Handling dates
In the given csv files date's datatype is string so you can not fetch its year ,month or anything , to overcome this we use this 

pd.read_csv("aug_test.csv",parse_dates=["date"])

# Convertors
This feature is used where suppose you have a column and it contains very long name and you want short name for them then you can use convertors.

pd.read_csv("aug_test.csv",convertors={"column name":function name})

# na_values parameters
This is used for missing values sometimes in the place of missing values there would be placeholders like (-,%,;) so we can use na_values it convers all these into the NaN .


pd.read_csv("aug_test.csv",na_values=['-','%'])

# Loading a huge dataset in chunks
