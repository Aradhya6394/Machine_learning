**Salary Prediction using Years of Experience**

```python
# import library
import pandas as pd
import numpy as np
```

```python
# load dataset
df=pd.read_csv("Salary_Data.csv")
```

```python
# read first five line
df.head()
```

```python
# find the information about the data
df.info()
```

```python
# find the shape
df.shape
```

```python
# find the statistics about the data
# we can find the outliyer with the help of statistics.
df.describe()
```

```python
import seaborn as sns
```

```python
sns.boxplot(df['Salary'])
```

```python
sns.boxplot(df['YearsExperience'])
```

```python
# divide x and y column
x=df[['YearsExperience']]
y=df['Salary']
```

```python
print(x)
```

```python
print(y)
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
# import matplotlib to visualize the dataset
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y,color='red')
plt.title('Experience vs Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
# we find that the dataset is linear
```

**Model Training**

```python
# import the library to train the model
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(y_pred)
```

```python
m=model.coef_[0]
c=model.intercept_
```

```python
print(m)
print(c)
```

```python
# draw the graph between y_pred and y_test
plt.scatter(x_test,y_test,color='blue',label='Actual')
plt.scatter(x_test,y_pred,color='red',label='Predicted')
plt.title('Actual vs Predicted test set')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()  # this shows the labels
plt.show()
```

```python
# draw the graph between x_train and y_train
plt.scatter(x_train,y_train,color='blue',label='Actual')
plt.scatter(x_train,model.predict(x_train),color='red',label='Predicted')
plt.title('Actual vs Predicted train set')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()  # this shows the labels
plt.show()
```

**Model Evaluation**

```python
# for model evaluation we calculate the MSE,RMSE,R2 SCORE
```

```python
# import requried library to calculate MSE,RMSE,R2  score
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
rmse=np.sqrt(mse)
```

```python
print("mse : ",mse)
print("r2 : ",r2)
print("rmse : ",rmse)
```
