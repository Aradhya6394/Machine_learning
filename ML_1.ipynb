```python
import numpy as np
from sklearn.datasets import load_iris
```

**Load the dataset**

```python

iris = load_iris()
```

```python
iris
```

```python
iris.data.shape
```

**selecting first 50 samples**

```python
x = iris.data[:50, 0]
y = iris.data[:50, 1]
```

```python
print(x)
```

```python
print(y)
```

**Here my x and y are in 1D row vector and for slop and intercept i want column vector so i reshape them.**

```python
x = x.reshape(-1, 1)  # converting into columns
y = y.reshape(-1, 1)
```

```python
print(x)
```

```python
print(y)
```

**Check shape of both x and y**

```python

x.data.shape
```

```python
y.data.shape
```

**Add columns for intercept**

```python
X = np.hstack((np.ones((x.shape[0], 1)), x))
```

**Matrix Method**

```python

w = np.linalg.inv(X.T @ X) @ X.T @ y
```

**Intercept and slop .**

```python
intercept = w[0, 0]
slope = w[1, 0]
```

```python
print("Intercept:", intercept)
print("Slope:", slope)
```

**Plot between x and y**

```python

y_pred = slope * x + intercept
```

```python
print(y_pred)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Linear Regression on Iris Dataset")
plt.show()
```

**Formula Method**

```python

x1 = x.flatten()
y1 = y.flatten()

n = len(x1)

x_mean = np.mean(x1)
y_mean = np.mean(y1)

numerator = np.sum((x1 - x_mean) * (y1 - y_mean))
denominator = np.sum((x1 - x_mean) ** 2)

slope_f = numerator / denominator
intercept_f = y_mean - slope_f * x_mean

print("Normal Formula Method")
print("Slope:", slope_f)
print("Intercept:", intercept_f)
# Prediction
y_pred_f = slope_f * x1 + intercept_f

# Plot
plt.scatter(x1, y1)
plt.plot(x1, y_pred_f, color='red')
plt.title("Linear Regression - Normal Formula Method")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

**Gradient Decent**

```python


x_f = x.flatten()
y_f = y.flatten()

slope = 0
intercept = 0
learning_rate = 0.01
epochs = 1000
n = len(x_f)

for _ in range(epochs):
    y_pred = slope*x_f + intercept

    d_slope = (-2/n) * np.sum(x_f * (y_f - y_pred))
    d_intercept = (-2/n) * np.sum(y_f - y_pred)

    slope = slope - learning_rate * d_slope
    intercept = intercept - learning_rate * d_intercept

print("Intercept:", intercept)
print("Slope:", slope)

y_pred = slope * x + intercept

plt.scatter(x, y)
plt.plot(x, y_pred)
plt.show()
```

**Covarience Method**

```python


x_f = x.flatten()
y_f = y.flatten()

x_mean = np.mean(x_f)
y_mean = np.mean(y_f)

x_centered = x_f - x_mean
y_centered = y_f - y_mean

slope = np.dot(x_centered, y_centered) / np.dot(x_centered, x_centered)
intercept = y_mean - slope * x_mean

print("Intercept:", intercept)
print("Slope:", slope)

y_pred = slope * x + intercept

plt.scatter(x, y)
plt.plot(x, y_pred)
plt.show()
```

**Inbuilt Function Method**

```python


from sklearn.linear_model import LinearRegression

x_f = x.flatten()
y_f = y.flatten()

x_sklearn = x_f.reshape(-1, 1)

model = LinearRegression()
model.fit(x_sklearn, y_f)

slope = model.coef_[0]
intercept = model.intercept_

print("Intercept:", intercept)
print("Slope:", slope)

y_pred = model.predict(x_sklearn)

plt.scatter(x, y)
plt.plot(x, y_pred)
plt.show()
```
